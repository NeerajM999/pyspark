```python
# replace ? with nan across dataframe
df5 = df.na.replace('?', None)
df5.selectExpr("encounter_id", "race", "gender", "age", "weight", "payer_code", "medical_specialty")\
    .where("weight is null").show(4)
```

```python
# race, payer_code and medical_specialty have missing values, so assign 'UNKNOWN' to missing values
df6 = df5.na.fill("UNKNOWN")
df6.createOrReplaceTempView("df6")
```

```python
df6.select("race", "payer_code", "medical_specialty", "weight").show(5)
```

```python
# get unique value count for each medical_specialty
top10 =  df6.select("medical_specialty").groupBy("medical_specialty").count().orderBy("count",ascending=False).limit(10)
top10.show(20, False)
top10.createOrReplaceTempView("top10")
```

```python
# set anything not in top 10 categories as 'Others'
df7 = spark.sql("""
    select *,
    case when medical_specialty not in ('UNKNOWN','InternalMedicine','Emergency/Trauma','Cardiology','Surgery-General','Nephrology','Orthopedics','Orthopedics-Reconstructive','Radiologist') 
    then 'Others' else medical_specialty end as new_med
    from df6
""")

df7 = df7.drop('medical_specialty')
df7.select("race", "payer_code", "weight", "new_med").show()
```

```python
df7.groupBy("new_med").count().orderBy("count", ascending=False).show(10)
```
